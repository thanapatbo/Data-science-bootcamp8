```python
import pandas as pd
import requests
import time
```

```python
from requests import get

## create obj_list
char_name = []
char_house = []
char_wand = []

for i in range(21):
    url = "https://hp-api.onrender.com/api/characters/students"
    resp = requests.get(url)
    #check request
    if resp.status_code == 200:
        json_list = resp.json()
        char_name.append(json_list[i]["name"])
        char_house.append(json_list[i]["house"])
        char_wand.append(json_list[i]["wand"])
    else:
        char_name.append("error")
        char_house.append("error")
        char_wand.append("error")
    time.sleep(1)
```

```python
print(char_name)
```

```python
## DataFrame
info_char_hp = pd.DataFrame( data = {
    "Character" : char_name,
    "House" : char_house,
    "Wand" : char_wand
})
```
