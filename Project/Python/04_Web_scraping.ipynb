```python
!pip install gazpacho
## import function soup
from gazpacho import Soup
import requests
```

```python
## homework web scraping 101
url = "https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc"

html = requests.get(url)
```

```python
html.text
```

```python
imdb = Soup(html.text)
```

```python
titles = imdb.find("h3", {"class": "lister-item-header"})
```

```python
clean_titles = [title.strip() for title in titles]
clean_titles
```

```python
ratings = imdb.find("div", {"class": "inline-block ratings-imdb-rating"})
clean_ratings = [rate.strip() for rate in ratings]
clean_ratings
```

```python
years = imdb.find("span", {"class" : "lister-item-year text-muted unbold"})
clean_y_relea = [year.strip() for year in years]
clean_y_relea
```

```python
## import pandas
import pandas as pd
```

```python
movies_data = pd.DataFrame(data={
    "Title" : clean_titles,
    "Rating" : clean_ratings,
    "Release Year" : clean_y_relea
})
```
